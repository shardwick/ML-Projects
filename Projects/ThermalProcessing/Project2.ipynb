# Predicting Motor Vehicle Temperature using Thermal Models

The goal of this project is to develop a machine learning model that can predict the temperature of a motor vehicle based on various thermal parameters. The dataset used in this project consists of historical thermal data collected from motor vehicles, including ambient temperature, engine temperature, radiator temperature, and other relevant features.

```python
# imports 
import random
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# Define function to simulate car data
def simulate_car_data():
    ambient_temperature = np.random.uniform(50, 104)  # Simulating ambient temperature (in Fahrenheit)
    engine_temperature = np.random.uniform(158, 230)  # Simulating engine temperature (in Fahrenheit)
    radiator_temperature = np.random.uniform(140, 194)  # Simulating radiator temperature (in Fahrenheit)
    vehicle_temperature = 77 + (0.036 * ambient_temperature) + (0.09 * engine_temperature) + (0.054 * radiator_temperature) + np.random.normal(0, 9)
    
  
    # Return the simulated car data as a dictionary
    car_data = {
        'Ambient Temperature': ambient_temperature,
        'Engine Temperature': engine_temperature,
        'Radiator Temperature': radiator_temperature,
        'Vehicle Temperature': vehicle_temperature
    }

    return car_data
```

```python

# Simulate car data samples
num_samples = 100
car_data_samples = []

for _ in range(num_samples * 10000):
    car_data = simulate_car_data()
    car_data_samples.append(car_data)



```

```python
# Convert car data to numpy arrays
X = np.array([[sample['Ambient Temperature'], sample['Engine Temperature'], sample['Radiator Temperature']] for sample in car_data_samples])
y = np.array([sample['Vehicle Temperature'] for sample in car_data_samples])


```

```python
# Fit a linear regression model
model = LinearRegression()
model.fit(X, y)

```

```python
# Make predictions
y_pred = model.predict(X)

```

```python
# Calculate model evaluation metrics
mae = mean_absolute_error(y, y_pred)
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

# Print model evaluation metrics
print("Model Evaluation Metrics:")
print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("R-squared (R2):", r2)

```

```python

# Plot actual vs. predicted vehicle temperature
plt.figure(figsize=(8, 6))
plt.scatter(y, y_pred, color='blue', alpha=0.6)
plt.plot([y.min(), y.max()], [y.min(), y.max()], color='red', linestyle='--')
plt.xlabel('Actual Vehicle Temperature (°F)')
plt.ylabel('Predicted Vehicle Temperature (°F)')
plt.title('Actual vs. Predicted Vehicle Temperature')
plt.show()

# Print model evaluation metrics
print("Model Evaluation Metrics:")
print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("R-squared (R2):", r2)
```
